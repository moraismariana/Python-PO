```python
#QUESTÃO 7
limite = int(input("Insira um número maior que 2: "))
lista = list(range(2, limite+1))
p = 2
while p < limite:
    for i in lista:
        if (i != p) and (i % p == 0):
            lista.remove(i)
    index = lista.index(p)
    if index + 1 < len(lista):
        p = lista[index + 1]
    else:
        break
print("A lista dos números primos entre 2 e {} é:".format(limite))
print(lista)
```

```python
#QUESTÃO 8
pontuacao = {
    "p1": ["a", "e", "i", "l", "n", "o", "r", "r", "s", "t", "u"],
    "p2": ["d", "g"],
    "p3": ["b", "c", "m", "p"],
    "p4": ["f", "h", "v", "w", "y"],
    "p5": ["k"],
    "p8": ["j", "x"],
    "p10": ["q", "z"],
}

palavra = input("Insira uma palavra: ").lower()
lista = list(palavra)
cont = 0

for i in lista:
    if i in pontuacao["p1"]:
        cont += 1
    elif i in pontuacao["p2"]:
        cont += 2
    elif i in pontuacao["p3"]:
        cont += 3
    elif i in pontuacao["p4"]:
        cont += 4
    elif i in pontuacao["p5"]:
        cont += 5
    elif i in pontuacao["p8"]:
        cont += 8
    elif i in pontuacao["p10"]:
        cont += 10

print("A pontuação dessa palavra é de {} pontos.".format(cont))
```

```python
#QUESTÃO 10
def calcular_distância(primeira, segunda):
    if len(primeira) == 0:
        return len(segunda)
    elif len(segunda) == 0:
        return len(primeira)
    else:
        custo = 0
        if primeira[-1] != segunda[-1]:
            custo = 1
        dist1 = calcular_distância(primeira[:-1], segunda) + 1
        dist2 = calcular_distância(primeira, segunda[:-1]) + 1
        dist3 = calcular_distância(primeira[:-1], segunda[:-1]) + custo

        return min(dist1, dist2, dist3)

string1 = input("Insira a primeira palavra: ")
string2 = input("Insira a segunda palavra: ")

print (f"A distância entre {string1} e {string2} é: {calcular_distância(string1, string2)}")
```
